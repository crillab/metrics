```python
from metrics.wallet import *
```

```python
analysis = Analysis(input_file='config/metrics_scalpel.yml').sub_analysis(
	column='Category', 
	sub_set={'CSP'}
)

available_xp_wares = set(analysis.campaign_df.xp_ware_names) - {'Concrete 3.12.2', 'cosoco 2.0'}
available_xp_wares_not_para = {x for x in available_xp_wares if 'parallel' not in x and 'Many' not in x}
available_xp_wares_para = {x for x in available_xp_wares if 'parallel' in x or 'Many' in x}
```

```python
analysis_not_para = analysis.sub_analysis(
	column='experiment_ware', 
	sub_set=available_xp_wares_not_para
).add_vbew(
	opti_col='cpu_time',
	minimize=True,
	vbew_name='VBS'
)
```

```python
analysis_not_para.get_cactus_plot()
```

```python
analysis_para = analysis.sub_analysis(
	column='experiment_ware', 
	sub_set=available_xp_wares_para
).add_vbew(
	opti_col='cpu_time',
	minimize=True,
	vbew_name='VBS'
)
```

```python
analysis_para.get_cactus_plot()
```
